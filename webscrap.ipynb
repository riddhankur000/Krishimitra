```python
# Web scraping using Selenium (example: clicking options & getting table data)

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Initialize Chrome WebDriver (make sure ChromeDriver is installed and PATH set)
driver = webdriver.Chrome()

try:
    # Step 1: Open the target website
    driver.get("https://enam.gov.in/web/dashboard/live_price")

    # Step 2: Wait for the page to load completely
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

    # Step 3: Click on the 'Commodity Wise' radio button
    commodity_radio = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//input[@value='commodity']"))
    )
    commodity_radio.click()

    # Step 4: Select a commodity from the dropdown menu (for example 'BAJRA')
    select_dropdown = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.TAG_NAME, 'select'))
    )
    select = Select(select_dropdown)
    select.select_by_visible_text('BAJRA')

    # Step 5: Wait for the dynamic table to appear
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//table"))
    )

    # Step 6: Extract the page source and parse it with BeautifulSoup
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Step 7: Locate and extract data from the table
    table = soup.find('table')
    rows = table.find_all('tr')

    print("\n=== Mandi Prices ===\n")
    for row in rows[1:]:  # skip header row
        cols = [td.text.strip() for td in row.find_all('td')]
        print(cols)

    # Optional: You can save this data to a CSV file if needed
    # import csv
    # with open('mandi_prices.csv', 'w', newline='', encoding='utf-8') as f:
    #     writer = csv.writer(f)
    #     writer.writerows([cols])

finally:
    time.sleep(5)
    driver.quit()
```
